```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2*np.pi*t)
s2 = np.sin(4*np.pi*t)
```

```python
plt.figure(1)
plt.subplot(211)
plt.ylabel('shit')
plt.xlabel('dick')
plt.plot(t, s1)
plt.ylabel('shit')
plt.xlabel('dick')
plt.subplot(212)
plt.plot(t, 2*s1)
```

```python
fig, (ax1, ax2,ax3) = plt.subplots(3)
fig.suptitle('Learning objectives')
ax1.set(xlabel='x-label', ylabel='y-label')
ax1.plot(t, s1)
ax2.set(xlabel='x-label', ylabel='y-label')
ax2.plot(t, 2*s1)
ax3.set(xlabel='x-label', ylabel='y-label')
ax3.plot(t, 2*s1)
fig.savefig('books_read.png')
```
